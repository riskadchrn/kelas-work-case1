```python
import pandas as pd
from scipy.stats import ttest_1samp
transaction_data = [
    100, 150, 50, 100, 130,
    120, 100, 85, 70, 150,
    150, 120, 50, 100, 100,
    140, 90, 150, 50, 90,
    120, 100, 110, 75, 65]
df = pd.DataFrame(transaction_data, columns=['TransactionAmount'])
print (df)
```

```python
df.describe()
```

```python
mean = df['TransactionAmount'].mean()
median = df['TransactionAmount'].median()
mode = df['TransactionAmount'].mode()[0]

print("Mean (Rata-rata): ", mean)
print("Median (Tengah): ", median)
print("Mode (Modus): ", mode)
```

```python
variance = df['TransactionAmount'].var()
std_deviation = df['TransactionAmount'].std()
range_value = df['TransactionAmount'].max() - df['TransactionAmount'].min()

print("Variance (Variansi): ", variance)
print("Standart Deviation (Simpangan baku): ", std_deviation)
print("Range (Jangkauan): ", range_value)
```

```python
t_stat, p_value = ttest_1samp(df['TransactionAmount'], 100, alternative='greater')
print("Hasil T-Test:")
print("T-Statistic:", t_stat)
print("p-Value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
  print("Keputusan: Tolak H0 (rata-rata penjualan meningkat signifikan setelah pelatihan)")
else:
  print("Keputusan: Gagal tolak H0 (tidak ada bukti rata-rata penjualan meningkat signifikan)")
```
